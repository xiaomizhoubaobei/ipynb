## LoRA 微调

LoRA 秩（rank）决定了添加到原始权重中的可训练矩阵维度，控制微调表达的细腻程度与精度。

- 秩越高 → 可捕捉更细粒度变化，但参数量与计算开销增大；  
- 秩越低 → 计算高效，但可能欠拟合。

本教程使用秩为 4。实际使用时建议先从小秩（4、8、16）开始实验，逐步提升并评估效果。  
注意避免过拟合/欠拟合：秩、学习率、训练步数等均需平衡。

启用 LoRA 后，可训练参数量从 26 亿锐减至 290 万，大幅降低显存与计算压力。
